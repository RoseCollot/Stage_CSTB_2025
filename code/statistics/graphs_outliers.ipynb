```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn.objects as so
from matplotlib.gridspec import GridSpec
from pandas.plotting import table 
```

```python

df_homo = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000001405.40/df_stats')
df_pan_tro = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_028858775.2/df_stats')
df_propi = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000956105.1/df_stats')
df_otolemur = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000181295.1/df_stats')
df_carlito = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000164805.1/df_stats')
df_colobus = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000951035.1/df_stats')
df_mandrillus = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000951045.1/df_stats')
df_control = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_003255815.1/df_stats')

# Create the figure and grid layout
fig = plt.figure(figsize=(12, 12))  # Adjust figure size
gs = GridSpec(4, 2, figure=fig)  # Create a 4x2 grid

# Add each plot to the grid
ax1 = fig.add_subplot(gs[0, 0])
sns.histplot(df_homo['ratio_introns'], ax=ax1)
ax1.set_title("Homo sapiens")
ax1.axvline(df_homo['ratio_introns'].mean(), color='k', lw=2)
ax1.axvline(df_homo['ratio_introns'].median(), color='k', ls='--', lw=2)

ax2 = fig.add_subplot(gs[0, 1])
sns.histplot(df_pan_tro['ratio_introns'], ax=ax2)
ax2.set_title("Pan troglodytes")
ax2.axvline(df_pan_tro['ratio_introns'].mean(), color='k', lw=2)
ax2.axvline(df_pan_tro['ratio_introns'].median(), color='k', ls='--', lw=2)

ax3 = fig.add_subplot(gs[1, 0])
sns.histplot(df_propi['ratio_introns'], ax=ax3)
ax3.set_title("Propithecus")
ax3.axvline(df_propi['ratio_introns'].mean(), color='k', lw=2)
ax3.axvline(df_propi['ratio_introns'].median(), color='k', ls='--', lw=2)

ax4 = fig.add_subplot(gs[1, 1])
sns.histplot(df_otolemur['ratio_introns'], ax=ax4)
ax4.set_title("Otolemur")
ax4.axvline(df_otolemur['ratio_introns'].mean(), color='k', lw=2)
ax4.axvline(df_otolemur['ratio_introns'].median(), color='k', ls='--', lw=2)

ax5 = fig.add_subplot(gs[2, 0])
sns.histplot(df_carlito['ratio_introns'], ax=ax5)
ax5.set_title("Carlito")
ax5.axvline(df_carlito['ratio_introns'].mean(), color='k', lw=2)
ax5.axvline(df_carlito['ratio_introns'].median(), color='k', ls='--', lw=2)

ax6 = fig.add_subplot(gs[2, 1])
sns.histplot(df_colobus['ratio_introns'], ax=ax6)
ax6.set_title("Colobus")
ax6.axvline(df_colobus['ratio_introns'].mean(), color='k', lw=2)
ax6.axvline(df_colobus['ratio_introns'].median(), color='k', ls='--', lw=2)

ax7 = fig.add_subplot(gs[3, 0])
sns.histplot(df_mandrillus['ratio_introns'], ax=ax7)
ax7.set_title("Mandrillus")
ax7.axvline(df_mandrillus['ratio_introns'].mean(), color='k', lw=2)
ax7.axvline(df_mandrillus['ratio_introns'].median(), color='k', ls='--', lw=2)

ax8 = fig.add_subplot(gs[3, 1])
sns.histplot(df_control['ratio_introns'], ax=ax8)
ax8.set_title("Theropithecus")
ax8.axvline(df_control['ratio_introns'].mean(), color='k', lw=2)
ax8.axvline(df_control['ratio_introns'].median(), color='k', ls='--', lw=2)

# Adjust layout
plt.tight_layout()
plt.show()


```

```python
species=[]

theropithecus = []
theropithecus.append("Theropithecus (control)")
theropithecus.append(df_control['ratio_introns'].mean())
theropithecus.append(df_control['ratio_introns'].median())
species.append(theropithecus)

homo = []
homo.append('Homo spiens')
homo.append(df_homo['ratio_introns'].mean())
homo.append(df_homo['ratio_introns'].median())
species.append(homo)

pan = []
pan.append("Pan troglodytes")
pan.append(df_pan_tro['ratio_introns'].mean())
pan.append(df_pan_tro['ratio_introns'].median())
species.append(pan)

propi = []
propi.append("Propithecus")
propi.append(df_propi['ratio_introns'].mean())
propi.append(df_propi['ratio_introns'].median())
species.append(propi)

otolemur = []
otolemur.append("Otolemur")
otolemur.append(df_otolemur['ratio_introns'].mean())
otolemur.append(df_otolemur['ratio_introns'].median())
species.append(otolemur)

carlito = []
carlito.append("Carlito")
carlito.append(df_carlito['ratio_introns'].mean())
carlito.append(df_carlito['ratio_introns'].median())
species.append(carlito)

colobus = []
colobus.append("Colobus")
colobus.append(df_colobus['ratio_introns'].mean())
colobus.append(df_colobus['ratio_introns'].median())
species.append(colobus)

mandrillus = []
mandrillus.append("Mandrillus")
mandrillus.append(df_mandrillus['ratio_introns'].mean())
mandrillus.append(df_mandrillus['ratio_introns'].median())
species.append(mandrillus)

df_mean_median = pd.DataFrame(species, columns=['species', 'mean', 'median'])
df_mean_median.to_csv('/home/collot/stage_git/figures/mean_median.csv')
```
