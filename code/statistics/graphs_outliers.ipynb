```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn.objects as so
from matplotlib.gridspec import GridSpec
from pandas.plotting import table 
```

```python
df_pan_tro_control = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_028858775.2/df_stats')
df_otolemur = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000181295.1/df_stats')
df_carlito = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000164805.1/df_stats')
df_homo = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000001405.40/df_stats')

# Create the figure and grid layout
fig = plt.figure(figsize=(12, 12))  
gs = GridSpec(2, 2, figure=fig)  
title_font = {'weight': 'normal','size': 16}

# Add each plot to the grid
ax1 = fig.add_subplot(gs[0, 0])
sns.histplot(df_pan_tro_control['ratio_introns'], ax=ax1)
ax1.set_title(r"$\it{Pan\ troglodytes}$", fontdict=title_font)
ax1.axvline(df_pan_tro_control['ratio_introns'].mean(), color='k', lw=2)
ax1.axvline(df_pan_tro_control['ratio_introns'].median(), color='k', ls='--', lw=2)
ax1.tick_params(axis='both', which='major', labelsize=18)
ax1.set_xlabel("Ratio of Introns", fontsize=16)
ax1.set_ylabel("Frequency", fontsize=16)

ax2 = fig.add_subplot(gs[0, 1], sharey=ax1)
sns.histplot(df_otolemur['ratio_introns'], ax=ax2)
ax2.set_title(r"$\it{Otolemur\ garnettii}$", fontdict=title_font)
ax2.axvline(df_otolemur['ratio_introns'].mean(), color='k', lw=2)
ax2.axvline(df_otolemur['ratio_introns'].median(), color='k', ls='--', lw=2)
ax2.tick_params(axis='both', which='major', labelsize=18)
ax2.set_xlabel("Ratio of Introns", fontsize=16)
ax2.set_ylabel("Frequency", fontsize=16)

ax3 = fig.add_subplot(gs[1, 0],sharey=ax1)
sns.histplot(df_carlito['ratio_introns'], ax=ax3)
ax3.set_title(r"$\it{Carlito\ syrichta}$", fontdict=title_font)
ax3.axvline(df_carlito['ratio_introns'].mean(), color='k', lw=2)
ax3.axvline(df_carlito['ratio_introns'].median(), color='k', ls='--', lw=2)
ax3.tick_params(axis='both', which='major', labelsize=18)
ax3.set_xlabel("Ratio of Introns", fontsize=16)
ax3.set_ylabel("Frequency", fontsize=16)

ax4 = fig.add_subplot(gs[1, 1],sharey=ax1)
sns.histplot(df_homo['ratio_introns'], ax=ax4)
ax4.set_title(r"$\it{Homo\ sapiens}$", fontdict=title_font)
ax4.axvline(df_homo['ratio_introns'].mean(), color='k', lw=2)
ax4.axvline(df_homo['ratio_introns'].median(), color='k', ls='--', lw=2)
ax4.tick_params(axis='both', which='major', labelsize=18)
ax4.set_xlabel("Ratio of Introns", fontsize=16)
ax4.set_ylabel("Frequency", fontsize=16)

# Adjust layout
plt.tight_layout()
plt.show()


```

```python
df_homo = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000001405.40/df_stats')
df_homo = df_homo.dropna()
df_homo["species"] = r"$\it{Homo\ sapiens}$"

df_pan = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_028858775.2/df_stats')
df_pan = df_pan.dropna()
df_pan["species"] = r"$\it{Pan\ troglodytes}$"

df_propi = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000956105.1/df_stats')
df_propi = df_propi.dropna()
df_propi["species"] = r"$\it{Propithecus\ coquereli}$"

df_otolemur = pd.read_csv('/home/collot/stage/collot/collot/out_stats/output_dataframes/GCF_000181295.1/df_stats')
df_otolemur = df_otolemur.dropna()
df_otolemur["species"] = r"$\it{Otolemur\ garnettii}$"

selected_species = pd.concat([df_homo, df_pan, df_propi, df_otolemur], ignore_index=True)

p = (so.Plot(selected_species, x="species", y="ratio_introns")
    .add(so.Dot(color="g"), so.Perc(10))
    .layout(size=(8, 5))
    .label(x="Species", y="Ratio of Introns"))

p.show()
```
